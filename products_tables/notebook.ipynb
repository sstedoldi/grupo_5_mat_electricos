```python
import pandas as pd
# import openpyxl
```

```python
mat_elec=pd.read_excel('Lista de Materiales Eléctricos.xlsx')
mat_elec.fillna(method='ffill', inplace=True)
mat_elec
```

```python
to_drop=["VENTA SIN IVA.1", "VENTA FINAL.1"]
mat_elec.drop(to_drop, axis=1, inplace=True)
mat_elec["VENTA SIN IVA"]=mat_elec["VENTA SIN IVA"]/134.25 # valor del dolar en ese momento
mat_elec["VENTA FINAL"]=mat_elec["VENTA FINAL"]/134.25 # valor del dolar en ese momento


columns_names= {
    'CATEGORIA':'category',
    'RUBRO' : 'subcategory',
    'DESCRIPCION': 'description',
    'MARCA': 'brand',
    'IVA': 'vat',
    'VENTA SIN IVA':'nonvatPrice',
    'VENTA FINAL': 'finalPrice'
}

mat_elec.rename(columns=columns_names, inplace=True)
mat_elec
```

# Categories

```python
categories=pd.DataFrame(mat_elec.category.unique())
categories.rename(columns={0:'category'}, inplace=True)
categories['category']=categories['category'].str.replace('.','')
categories["id"]=(categories.index + 1).astype(str)
categories=categories[['id','category']]
print(categories)
categories.to_csv("categories_table.csv", index=False)
```

```python
categories_dict=dict(zip(categories['category'],categories['id'] ))
```

# Subcategories

```python
subcategories=pd.DataFrame(mat_elec.subcategory.unique())
subcategories.rename(columns={0:'subcategory'}, inplace=True)
subcategories['subcategory']=subcategories['subcategory'].str.replace('.','')
subcategories["id"]=(subcategories.index + 1).astype(str)
subcategories=subcategories[['id','subcategory']]
print(subcategories)
# subcategories.to_csv("subcategories_table.csv") lo hago luego, agregando el id de category
```

```python
subcategories_dict=dict(zip(subcategories['subcategory'],subcategories['id'] ))
```

# Brands

```python
brands=pd.DataFrame(mat_elec.brand.unique())
brands.rename(columns={0:'brand'}, inplace=True)
brands['brand']=brands['brand'].str.replace('.','')
brands["id"]=(brands.index + 1).astype(str)
brands=brands[['id','brand']]
print(brands)
brands.to_csv("brands_table.csv", index=False)
```

```python
brands_dict=dict(zip(brands['brand'],brands['id'] ))
brands_dict
```

```python
# mat_elec["categories_id"]=
mat_elec['category']=mat_elec['category'].str.replace('.','')
mat_elec['category_id']=mat_elec['category'].map(categories_dict)

mat_elec['subcategory']=mat_elec['subcategory'].str.replace('.','')
mat_elec['subcategory_id']=mat_elec['subcategory'].map(subcategories_dict)
mat_elec['brand_id']=mat_elec['brand'].map(brands_dict)
mat_elec
```

```python
subcategories=mat_elec[['subcategory_id','subcategory','category_id']]
# print(subcategories)
subcategories=subcategories.drop_duplicates(['subcategory_id','subcategory','category_id']).reset_index(drop=True)
subcategories['subcategory']=subcategories['subcategory'].str.replace('.','')
print(subcategories)
subcategories.to_csv("subcategories_table.csv", index=False)
```

```python
mat_elec['name']=""
mat_elec['stock']=0
mat_elec['stock_min']=0
mat_elec['discount']=0

to_drop=["category", "subcategory","category_id"]
mat_elec.drop(to_drop, axis=1, inplace=True)

# columns order
columns_ord=['name','description','vat','nonvatPrice','discount',"subcategory_id",'brand_id','stock','stock_min']
mat_elec=mat_elec[columns_ord]
mat_elec
```

# Agrego columna de category y image

```python
#mat_elec['image']=
mat_elec_all = mat_elec.merge(subcategories,on='subcategory_id')
#
mat_elec_all = mat_elec_all.merge(categories,left_on='category_id',right_on='id')
```
